```python
import numpy as np
import matplotlib.pyplot as plt
from pic1d2v import *
```

```python
N = 128
rho = np.sin(np.linspace(0,2*2*np.pi, N+1))[:-1]+\
      np.sin(np.linspace(0,10*2*np.pi, N+1))[:-1]
expected = np.sin(np.linspace(0,2*2*np.pi, N+1))[:-1]/4+\
           np.sin(np.linspace(0,10*2*np.pi, N+1))[:-1]/100
plt.plot(rho);
```

```python
rhok = np.fft.fft(rho)
k = len(rho)*np.fft.fftfreq(len(rho))

phik = np.zeros_like(rhok)
phik[1:] = rhok[1:]/(k[1:]**2)
sol = np.real(np.fft.ifft(phik))
plt.plot(sol)
plt.plot(expected)
np.linalg.norm(expected-sol)
```

```python
np.max(np.abs(np.imag(np.fft.ifft(phik))))
```

```python
A = one_d_poisson(N-1)
x = np.zeros(N)
x[1:] = np.linalg.solve(A, -((2*np.pi/N)**2)*rho[1:])
plt.plot(x)
plt.plot(expected)
np.linalg.norm(expected-x)
```
