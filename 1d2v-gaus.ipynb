```python
import numpy as np
from scipy.stats import norm
from pic1d2v import *
import matplotlib.pyplot as plt
from matplotlib import animation
from collections import namedtuple
```

```python
nt = 400
nx = 512
L  = 2*np.pi
dx = L/nx
dt = 0.1
B0 = 0

N_e = 2**10
N_i = 2**10
n0  = np.linspace(0., L, N_e+1)[:-1]
vx0 = .1*norm.pdf(n0, L*3/4, .25)
vx1 = .1*norm.pdf(n0, L*1/4, .25)

electron = Species(1., 1., N_e, n0,
                   vx0-1., np.zeros(N_e))
ion = Species(1., 1., N_i, n0,
                   vx1+1., np.zeros(N_i))
species = [electron, ion]
colors  = ['b', 'r']

xp, vx, vy, E, phi = pic(species, nx, dx, nt, dt, L, B0, 
                         solver_method="FFT",
                         weight_method="CIC",
                         interp_method="CIC")
```

```python
plt.plot(n0, vx0)
plt.plot(n0, vx1)
```

```python
fig = plt.figure()
p1 = fig.add_subplot(1, 1, 1)
#p2 = fig.add_subplot(1, 2, 2)
imgs = []   
s = 1
x_vals = np.linspace(0,L,nx+1)[:-1]
for i in range(nt+1):
    imgs.append((p1.scatter(xp[i,:N_e], vx[i,:N_e], color=colors[0], s=s),
                 p1.scatter(xp[i,N_e:], vx[i,N_e:], color=colors[1], s=s),
                 #p1.plot(x_vals, phi[i], color='g')[0],
                ))


im_ani = animation.ArtistAnimation(fig, imgs, 
                                   interval=50, repeat_delay=3000,
                                   blit=True)
im_ani.save("1d2v-gaus.mp4", writer="mencoder")
```
