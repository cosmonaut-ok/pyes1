```python
import numpy as np
from pic1d2v import *
import matplotlib.pyplot as plt
from matplotlib import animation
from collections import namedtuple
```

```python
nt = 200
nx = 512
L  = 2*np.pi
dx = L/nx
dt = 0.1
B0 = 0

N_e = 2**11
N_i  = N_e
n0 = np.linspace(0., L, N_e+1)[:-1]
n1 = .1*np.cos(n0/L*4*np.pi)
init_pos = n0 + n1
n0i  = np.linspace(0., L, N_i+1)[:-1]
normalize(init_pos, L)

electron = Species(-1., 1., N_e,
                   init_pos,
                   np.zeros(N_e), np.zeros(N_e))
ion      = Species(1., -1., N_i,
                   n0i,
                   np.zeros(N_i), np.zeros(N_i))

species = [electron, ion]
colors  = ['b']

xp, vx, vy, E, phi = pic(species, nx, dx, nt, dt, L, B0, 
                         solver_method="FFT",
                         weight_method="CIC",
                         interp_method="CIC")


```

```python
q_vals = np.zeros_like(electron.x0)
q_vals[:] = electron.q
wx = weight(electron.x0, q_vals, nx, L, method="CIC")
plt.plot(poisson_solve_fft(wx, dx))
wx = weight(electron.x0, q_vals, nx, L, method="NGP")
plt.plot(poisson_solve_fft(wx, dx))
```

```python
fig = plt.figure()
p1 = fig.add_subplot(1, 1, 1)
#p2 = fig.add_subplot(1, 2, 2)
imgs = []   
s = 1
x_vals = np.linspace(0,L,nx+1)[:-1]
for i in range(nt+1):
    imgs.append((p1.scatter(xp[i,:N_e], vx[i,:N_e], color=colors[0], s=s),
                 p1.plot(x_vals, phi[i], color='g')[0],
                ))


im_ani = animation.ArtistAnimation(fig, imgs, 
                                   interval=50, repeat_delay=3000,
                                   blit=True)
im_ani.save("1d2v-wave.mp4", writer="mencoder")
```
