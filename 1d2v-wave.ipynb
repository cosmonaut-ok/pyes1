```python
import numpy as np
from pic1d2v import *
import matplotlib.pyplot as plt
from matplotlib import animation
from collections import namedtuple
norm = lambda x: np.max(np.abs(x))
```

```python
nt = 150
nx = 128
mode = 1
L  = 2*np.pi
dx = L/nx
dt = 0.2
B0 = 0

N_e = 2**8
N_i  = N_e
x0 = np.linspace(0., L, N_e+1)[:-1]
n0 = N_e/L
k = mode*2*np.pi/L
x1 = .01*np.cos(x0*k)
init_pos = x0 + x1
x0i  = np.linspace(0., L, N_i+1)[:-1]
normalize(init_pos, L)

wp2 = 1.
q = wp2/n0
m = q

electron = Species(-q, m, N_e,
                   init_pos,
                   np.zeros(N_e), np.zeros(N_e))
ion      = Species(+q, -1., N_i,
                   x0i,
                   np.zeros(N_i), np.zeros(N_i))

species = [electron, ion]
colors  = ['b']

xp, vx, vy, E, phi = pic(species, nx, dx, nt, dt, L, B0, 
                         solver_method="FFT",
                         weight_method="CIC",
                         interp_method="CIC")
wp = np.sqrt(wp2)
print 'wp:', wp, 'q: ', q, 'm: ', m 
print L/N_e
```

```python
rho = np.zeros_like(phi)
qa = np.zeros_like(xp[0])
qa[:N_e] = electron.q
qa[N_e:] = ion.q
for i in range(nt+1):
    rho[i] = weight(xp[i], qa, nx, L, method="CIC")
plt.plot(rho[0])
```

```python
xpe = xp[:,:N_e]
k_vals = np.zeros((nt, nx/2+1), dtype=np.complex)
q_vals = np.zeros(N_e)
q_vals[:] = -q
rho0 = weight(x0, q_vals, nx, L, method="CIC")
for i in range(nt):
    rho = weight(xpe[i], q_vals, nx, L, method="CIC")
    
    rho_vals = rho-rho0
    k_vals[i] = np.fft.rfft(rho_vals) 

kn = np.sqrt(np.real(k_vals[:,mode]*k_vals[:,mode].conj()))
#kn = np.real(k_vals[:,mode])
plt.plot(kn)
plt.plot(np.abs(np.cos(wp*np.linspace(0,dt*nt, nt+1)[:-1])*np.max(kn)))
```

```python
i = 100

xt = (xp[:, :N_e]-x0)
t_vals = np.linspace(0,dt*nt, nt+1)
errors = [norm(xt[:,i]-np.cos(wp*t_vals)*xt[:,i][0]) for i in range(N_e)]

plt.plot(errors[1:])
```
