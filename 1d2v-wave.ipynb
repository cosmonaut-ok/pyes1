```python
import numpy as np
from pic1d2v import *
from picwave import *
import matplotlib.pyplot as plt
from matplotlib import animation
from collections import namedtuple
norm = lambda x: np.max(np.abs(x))
```

```python
pw = PicWave()
pw.run()

print 'wp:', pw.wp, 'q: ', pw.q, 'm: ', pw.m 
print pw.L/pw.N_e
```

```python
plt.plot(pw.rho[0])
```

```python
xpe = pw.xp[:,:pw.N_e]
k_vals = np.zeros((pw.nt, pw.nx/2+1), dtype=np.complex)
q_vals = np.zeros(pw.N_e)
q_vals[:] = -pw.q
rho0 = weight(pw.x0, q_vals, pw.nx, pw.L, method="CIC")
for i in range(pw.nt):
    rho = weight(xpe[i], q_vals, pw.nx, pw.L, method="CIC")
    
    rho_vals = rho-rho0
    k_vals[i] = np.fft.rfft(rho_vals) 

kn = np.sqrt(np.real(k_vals[:,pw.mode]*k_vals[:,pw.mode].conj()))
expected = np.abs(np.cos(pw.wp*np.linspace(0, pw.dt*pw.nt, 
                                        pw.nt+1)[:-1])*np.max(kn))
plt.plot(kn)
plt.plot(expected)
print norm(expected - kn)
```

```python
i = 100

xt = (pw.xp[:, :pw.N_e]-pw.x0)

t_vals = np.linspace(0,pw.dt*pw.nt, pw.nt+1)
errors = [norm(xt[:,i]-np.cos(pw.wp*t_vals)*xt[:,i][0]) 
          for i in range(pw.N_e)]

# Skip the first and last particle to avoid dealing with wrap around logic
plt.plot(errors[1:-1])
```
