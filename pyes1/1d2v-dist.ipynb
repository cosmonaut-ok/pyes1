```python
import numpy as np
from scipy.stats import norm
from pic1d2v import *
import matplotlib.pyplot as plt
from matplotlib import animation
from collections import namedtuple
```

```python
nt = 200
nx = 512
L  = 2*np.pi
dx = L/nx
dt = 0.1
B0 = 0

N_e1 = 2**9
N_e2 = 2**9
N_i  = N_e1+N_e2
n0e  = np.linspace(0., L, N_e1+1)[:-1]
n0i  = np.linspace(0., L, N_i+1)[:-1]
vx0 = .01*norm.rvs(L*3/4,  .1, N_e1)
vx1 = .01*norm.rvs(-L*1/4, .1, N_e2)
offset = 1.

q = .3
electron1 = Species(-q, 1., N_e1, n0e,
                    vx0-offset, np.zeros(N_e1))
electron2 = Species(-q, 1., N_e2, n0e,
                    vx1+offset, np.zeros(N_e2))
ion       = Species(q, -1., N_i, n0i,
                    np.zeros(N_i), np.zeros(N_i))
species = [electron1, electron2, ion]
colors  = ['b', 'r']

xp, vx, vy, E, phi, rho = pic(species, nx, dx, nt, dt, L, B0, 
                              solver_method="FFT",
                              weight_method="CIC",
                              interp_method="CIC")
```

```python
plt.plot(n0e, vx0-offset)
plt.plot(n0e, vx1+offset)
```

```python
fig = plt.figure()
p1 = fig.add_subplot(1, 1, 1)
imgs = []   
s = 1
x_vals = np.linspace(0,L,nx+1)[:-1]
for i in range(nt+1):
    imgs.append((p1.scatter(xp[i,:N_e1], vx[i,:N_e1], color=colors[0], s=s),
                 p1.scatter(xp[i,N_e1:N_e1+N_e2], 
                            vx[i,N_e1:N_e1+N_e2], color=colors[1], s=s),
                ))


im_ani = animation.ArtistAnimation(fig, imgs, 
                                   interval=50, repeat_delay=3000,
                                   blit=True)
im_ani.save("1d2v-dist.mp4", writer="mencoder")
```
