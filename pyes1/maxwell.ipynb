```python
import numpy as np
import scipy as sp
from scipy.stats import maxwell, norm
import matplotlib.pyplot as plt
```

```python
x_vals = np.linspace(0, 5, 100)
plt.plot(maxwell.pdf(x_vals, scale=1))
plt.plot(maxwell.pdf(x_vals, scale=2))
```

```python
plt.plot(x_vals, maxwell.pdf(x_vals, scale=1))
plt.hist(maxwell.rvs(size=10000), normed=True);
```

```python
x_vals = np.linspace(-5, 5, 100)
plt.plot(norm.pdf(x_vals, scale=1))
plt.plot(norm.pdf(x_vals, scale=2))
```

```python
KT = 2.
m  = 1.
scale = np.sqrt(KT/m)

expected = np.sqrt(m/(2*np.pi*KT))*np.exp(-m*x_vals**2/(KT*2))
plt.plot(x_vals, expected)
plt.plot(x_vals, norm.pdf(x_vals, scale=scale))

plt.hist(norm.rvs(scale=scale, size=10000), normed=True);
```
