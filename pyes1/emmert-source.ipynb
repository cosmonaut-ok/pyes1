```python
import numpy as np
import matplotlib.pyplot as plt
import scipy
from scipy.stats import rayleigh, binom
```

```python
def S(x, v, m, KT, h=lambda x: 1.0):
    return .5*h(x)*(m*np.abs(v)/KT)*np.exp(-m*v**2/(2*KT))
```

```python
m  = 1.0
KT = 1.0

scale = np.sqrt(KT/m)
v_vals = np.linspace(-4.0, 4.0, 1000)
```

```python
plt.plot(v_vals, S(0, v_vals, m, KT))
plt.plot(v_vals, .5*rayleigh.pdf(np.abs(v_vals), scale=scale))
```

```python
size = 1e6
d = binom.rvs(1, .5, size=size)
d[d==0] = -1
d[d==1] = 1
sample  = d*rayleigh.rvs(scale=scale, size=size)

plt.hist(sample, normed=True, bins=30)
plt.plot(v_vals, S(0, v_vals, m, KT))
plt.xlabel(r"$v_x$")
plt.ylabel(r"$S$")
```
